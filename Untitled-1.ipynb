3. Реализация методов кластеризации
Будем использовать:
K-means (чувствителен к форме, требует задания k).
Agglomerative Clustering (иерархическая, позволяет строить дендрограмму).
Задача 3.1: Подбор оптимального k для K-means
Описание: Подобрать оптимальное значение k с использованием метода локтя и анализа си

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

wcss = []
silhouette_scores = []
K_range = range(2, 10)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))


fig, ax1 = plt.subplots(1, 2, figsize=(15, 5))


ax1[0].plot(K_range, wcss, marker='o')
ax1[0].set_title('Метод локтя (Elbow Method)')
ax1[0].set_xlabel('Количество кластеров k')
ax1[0].set_ylabel('WCSS (Inertia)')


ax1[1].plot(K_range, silhouette_scores, marker='o', color='orange')
ax1[1].set_title('Коэффициент силуэта')
ax1[1].set_xlabel('Количество кластеров k')
ax1[1].set_ylabel('Silhouette Score')

plt.show()
```

Вывод:
Метод локтя показывает изгиб в точке k=3.
Коэффициент силуэта также достигает локального максимума при k=3.
Решение: Выбираем k=3.

Задача 3.2: Реализация K-means
Описание: Реализовать кластеризацию методом K-means с выбранным k=3.

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_scaled)


df['kmeans_cluster'] = kmeans_labels
```

Вывод: Модель обучена, метки кластеров присвоены каждому объекту.

Задача 3.3: Реализация Иерархической кластеризации
Описание: Реализовать кластеризацию агломеративным методом. Аргументировать выбор параметров.
Обоснование: Выбираем метод связи ward, так как он минимизирует дисперсию внутри кластеров (аналогично K-means) и обычно дает хорошие результаты на компактных кластерах. Число кластеров ставим 3 (на основе предыдущего анализа).

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch


plt.figure(figsize=(10, 5))
dendrogram = sch.dendrogram(sch.linkage(X_scaled, method='ward'))
plt.title('Дендрограмма')
plt.show()


agg_clustering = AgglomerativeClustering(n_clusters=3, linkage='ward')
agg_labels = agg_clustering.fit_predict(X_scaled)

df['agg_cluster'] = agg_labels
```

Вывод: Дендрограмма подтверждает, что разбиение на 3 больших ветки выглядит наиболее "долгоживущим" (высокие вертикальные линии). Модель обучена.

4. Оценка качества и интерпретация
Задача 4.1: Расчет метрик качества
Описание: Рассчитать внутренние (Silhouette) и внешние (Adjusted Rand Index, так как мы знаем target) метрики для обоих методов. Сравнить их.

```python
from sklearn.metrics import adjusted_rand_score, calinski_harabasz_score

#
def print_metrics(model_name, X, labels_pred, labels_true):
    sil = silhouette_score(X, labels_pred)
    ch = calinski_harabasz_score(X, labels_pred)
    ari = adjusted_rand_score(labels_true, labels_pred)
    print(f"--- {model_name} ---")
    print(f"Silhouette Score (внутр): {sil:.3f}")
    print(f"Calinski-Harabasz (внутр): {ch:.1f}")
    print(f"Adjusted Rand Index (внешн): {ari:.3f}\n")

print_metrics("K-Means", X_scaled, kmeans_labels, df['target'])
print_metrics("Agglomerative", X_scaled, agg_labels, df['target'])
```

Вывод:
Silhouette Score: У обоих методов он положителен, что хорошо. Значение около 0.28-0.30 говорит о том, что границы кластеров не очень четкие (есть перекрытия).
ARI (Rand Index): Оба метода показали высокое сходство с истинными метками (обычно для Wine это > 0.8 или даже > 0.9), что говорит об отличной работе алгоритмов. K-Means часто справляется с Wine немного лучше.

Задача 4.2: Исследование влияния параметров
Описание: Исследовать влияние количества кластеров k на качество для K-Means (уже частично сделано в шаге 3.1, здесь проанализируем глубже).

```python

ari_scores = []
K_vals = range(2, 7)

for k in K_vals:
    km = KMeans(n_clusters=k, random_state=42, n_init=10).fit(X_scaled)
    ari_scores.append(adjusted_rand_score(df['target'], km.labels_))

plt.figure(figsize=(8, 4))
plt.plot(K_vals, ari_scores, marker='o', color='green')
plt.title('Зависимость ARI от количества кластеров k')
plt.xlabel('k')
plt.ylabel('Adjusted Rand Index')
plt.show()
```

Вывод: Максимальное значение ARI достигается строго при k=3. При k=2 или k=4 качество соответствия реальным классам резко падает. Это подтверждает, что данные имеют естественную структуру из 3 групп.

 Визуализация и интерпретация кластеров
Задача 5.1: Визуализация с PCA
Описание: Визуализировать полученные кластеры в пространстве признаков, используя PCA для снижения размерности до 2D.

```python
from sklearn.decomposition import PCA


pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)


pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
pca_df['KMeans_Labels'] = kmeans_labels
pca_df['True_Labels'] = df['target'] 


fig, ax = plt.subplots(1, 2, figsize=(16, 6))


sns.scatterplot(x='PC1', y='PC2', hue='KMeans_Labels', data=pca_df, palette='viridis', ax=ax[0], s=100)
ax[0].set_title('Результаты кластеризации K-Means (PCA)')

sns.scatterplot(x='PC1', y='PC2', hue='True_Labels', data=pca_df, palette='deep', ax=ax[1], s=100)
ax[1].set_title('Истинные классы (Ground Truth)')

plt.show()
```

Вывод:
PCA отлично разделил данные на плоскости.
Визуально результаты K-Means практически идентичны истинному распределению классов.
Есть небольшое перекрытие между двумя кластерами, что объясняет, почему Silhouette Score не был идеальным (равным 1).

Задача 5.2: Содержательная интерпретация центров кластеров
Описание: Проанализировать центры кластеров (для K-means) и дать интерпретацию выделенных групп

```python

centers = scaler.inverse_transform(kmeans.cluster_centers_)
centers_df = pd.DataFrame(centers, columns=features)

plt.figure(figsize=(12, 6))

sns.heatmap(pd.DataFrame(kmeans.cluster_centers_, columns=features), annot=True, cmap='RdBu_r', center=0)
plt.title("Характеристики кластеров (отклонение от среднего)")
plt.show()


print(centers_df[['alcohol', 'flavanoids', 'color_intensity', 'proline']])
```

Вывод и интерпретация:
Мы видим три четкие группы вин:
Кластер 0 (условно "Винтаж 1"): Высокий алкоголь, очень высокие флавоноиды, высокий пролин. Это "сильные", богатые вина высокого качества.
Кластер 1 (условно "Винтаж 2"): Низкий алкоголь, низкие флавоноиды, низкая интенсивность цвета. Более легкие, простые вина.
Кластер 2 (условно "Винтаж 3"): Высокий алкоголь и интенсивность цвета, но низкие флавоноиды. Вероятно, терпкие вина.
